## Practicing File Handling Questions

### Assignment 1 : Reading a File
Write a function that reads the contents of a file named `sample.txt` and prints each line.

```python
def read_file(filename):
    with open(filename,'r') as file:
        for line in file:
            print(line.strip())
read_file('sample.txt')
```

### Assignment 2 : Writing to a File

Write a function that writes a list of strings to a file named `output.txt`, with each string on a new line.

```python
def write_list(filename,lst):
    with open(filename,'w') as file:
        for line in lst:
            file.write(line)
write_list('output.txt',['This is a very friendly language','This is easy to learn'])
```

### Assignment 3 : Copying a file
Write a function that copies the contents of a file named `sample.txt` to a new file named `destination.txt`.

```python
def copy_content(sourcefile,destinationfile):
    with open(sourcefile,'r') as file:
        contents=file.read()
    with open(destinationfile,'w') as file:
        file.write(contents)
copy_content('sample.txt','destination.txt')
```

### Assignment 4 : Appending to a File

Write a function that appends a given string to the end of a file named `sample.txt`.

```python
def append_string(filename,string):
    with open(filename,'a') as file:
        file.write(string)
append_string('sample.txt','\nAppending data to the end of a file.')
```

### Assignment 5: Counting Words in a File

Write a function that reads the contents of a file named `sample.txt` and returns the number of words in the file.

```python
def count_words(filename):
    with open(filename,'r') as file:
        contents=file.read()
        return len(contents.split())
count_words('sample.txt')
```


### Assignment 6: Finding and Replacing Text

Write a function that finds and replaces all occurrences of a given word in a file named `sample.txt` with another word.


```python
def find_and_replace(filename,to_find,to_replace_with):
    with open(filename,'r+') as file:
        contents=file.read()
        contents=contents.replace(str(to_find),str(to_replace_with))

        file.seek(0)
    
        file.write(contents)
        file.truncate()
    return contents
print(find_and_replace('sample.txt','3.10','3.13'))
```

### Assignment 7: Reading a File in Reverse

Write a function that reads the contents of a file named `sample.txt` and prints each line in reverse order.

```python
def reading_in_reverse(filename):
    with open(filename,'r') as file:
        contents=file.read()
        contents=contents.split('\n')
        contents=contents[::-1]
    for i in range(len(contents)):
        print(contents[i])
reading_in_reverse('sample.txt')
```

### Assignment 8: Counting Lines, Words, and Characters

Write a function that reads the contents of a file named `sample.txt` and returns the number of lines, words, and characters in the file.

```python
def number(filename):
    with open(filename,'r') as file:
        contents=file.read()
        n_lines=len(contents.split('\n'))
        n_words=len(contents.split()) 
        n_characters=len(contents)
    return n_lines,n_words,n_characters
n_lines,n_words,n_characters=number('sample.txt')
print(n_lines,n_words,n_characters)
```

### Assignment 9: Merging Multiple Files

Write a function that merges the contents of multiple files into a single file named `merged.txt`.


```python
def merge_files(file_list,merged_file_name):
    with open(merged_file_name,'w') as outfile:
        for file_name in file_list:
            with open(file_name,'r') as infile:
                outfile.write(infile.read())
merge_files(['output.txt','destination.txt','sample.txt'],'merged.txt')
            
```

### Assignment 10: Splitting a Large File

Write a function that splits a large file named `large.txt` into smaller files of 100 lines each.

```python
def split_file(filename):
    with open(filename,'r') as file:
        lines=file.readlines()
    for i in range(0,len(lines),100):
        with open(f'file{i//100+1}.txt','w') as new_file:
            new_file.writelines(lines[i:i+100])
split_file('large.txt')
```
