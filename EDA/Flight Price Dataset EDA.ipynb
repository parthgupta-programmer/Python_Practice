```python
source_cities=df['source_city'].unique()
destination_cities=df['destination_city'].unique()

print(source_cities)
print(destination_cities)

print(f"\n{'Source City':<15}{'Destination City':<20}{'Avg Price':<12}\n")
for sc in source_cities:
    for dc in destination_cities:
        if sc != dc:
            temp = df[(df['source_city'] == sc) & (df['destination_city'] == dc)]
            print(f"{sc:<15}{dc:<20}{temp['price'].mean():<12.2f}")
    print()
```
