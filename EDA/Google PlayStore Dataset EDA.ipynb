# EDA and Feature Engineering on Google PlayStore Dataset 

## Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

df=pd.read_csv(r'googleplaystore.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
dfc=df.copy()
```

```python
dfc=dfc.drop(dfc.index[10472])
```

```python
dfc[~dfc['Reviews'].str.isnumeric()]
```

```python
dfc['Reviews']=dfc['Reviews'].astype(int)
```

```python
dfc.head()
```

```python
dfc['Size'].str[-1].unique()
```

```python
dfc['Size']=dfc['Size'].str.replace('M','000')
dfc['Size']=dfc['Size'].str.replace('k','')
dfc['Size']=dfc['Size'].replace('Varies with device',np.nan)
```

```python
dfc['Size'].unique()
```

```python
dfc['Size']=dfc['Size'].astype(float)

dfc.info()
```

```python
dfc.head()
```

```python
dfc['Installs'].unique()
```

```python
dfc['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']

for item in chars_to_remove:
    for cols in cols_to_clean:
        dfc[cols]=dfc[cols].replace(item,'')
```

```python
dfc.head()
```

```python
dfc['Price'].unique()
```

```python
dfc['Price']=dfc['Price'].astype(float)
dfc['Installs']=dfc['Installs'].astype(int)
```

```python
dfc.info()
```

```python
dfc.head()
```

```python
dfc['Last Updated'].unique()
```

```python
dfc['Last Updated']=pd.to_datetime(dfc['Last Updated'])

dfc['Day']=dfc['Last Updated'].dt.day
dfc['Month']=dfc['Last Updated'].dt.month
dfc['Year']=dfc['Last Updated'].dt.year
```

```python
dfc.drop(['Last Updated'],axis=1,inplace=True)
```

```python
dfc.head()
```

```python
dfc.to_csv('google_cleaned.csv')
```

```python
numeric_feature=[feature for feature in dfc.columns if dfc[feature].dtype != 'O'] # O- Object
categorical_feature=[feature for feature in dfc.columns if dfc[feature].dtype == 'O']

print(f'Numerical Features: {numeric_feature}')
print(f'Categorical Features: {categorical_feature}')
```

```python
for column in categorical_feature:
    print(f'{column}:\n')
    print(f'{dfc[column].value_counts()}') # If in value_counts(noramlize=True)*100 # It will show the percentage for each category.
    print()
```

```python
## Proportion of count data on Numerical Columns

import seaborn as sns

plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(len(numeric_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=dfc[numeric_feature[i]],fill=True,color='r')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

```python
## Proportion of count data on Numerical Columns

import seaborn as sns

plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(len(numeric_feature)):
    plt.subplot(5,3,i+1)
    sns.histplot(x=dfc[numeric_feature[i]],color='r')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
plt.suptitle("Univariate Analysis of Categorical Features",fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(dfc[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
